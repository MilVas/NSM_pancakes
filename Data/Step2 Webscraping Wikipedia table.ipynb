```python
import pandas as pd
import requests
import bs4
```

```python
url = "https://en.wikipedia.org/wiki/Wikipedia:Multiyear_ranking_of_most_viewed_pages#Books_and_book_series"
```

```python
resp = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(resp.content, "html.parser")
```

```python
soup.body
```

```python
#Specific table: Books and book series
books_table = soup.find_all("table",{"class":"wikitable"})[17]
print(books_table)
```

```python
#Gets a list of the column names of the table
def get_columns(column):
    return [x.text.replace("\n","") for x in books_table.find_all("th")]
```

```python
column_names= get_columns(books_table)
print(column_names)
```

```python
#Gets a list of all the values of the table
def get_all_values(values):
    return [x.text.replace("\n","") for x in books_table.find_all("td")]
```

```python
elements = get_all_values(books_table)
print(elements)
```

```python
#Gets a list of the ranking of the books
def get_rank(rank): 
    return elements[::3]
    
get_rank(elements)
```

```python
#Gets a list of the titles
def get_title(title):
    return elements[1::3]

get_title(elements)
```

```python
#Gets a list of the views (million)
def get_views(views): 
    return elements[2::3]
    
get_views(elements)
```

```python
#Function that calls the ranking, the title, and the views functions and returns a dictionary.
def create_rows(items):
    return{"Rank" : get_rank(elements),
           "Page" : get_title(elements),
           "Views in millions" : get_views(elements)}

dict_values =dict(create_rows(elements))
print(dict_values)
```

```python
#Create the Data Frame
wiki_books_df = pd.DataFrame.from_dict(dict_values)
wiki_books_df.head(50)
```
